```python
def predict(paras, X):
    A2, cache = forward_propagation(X, paras)
    predictions = np.round(A2)
    
    return predictions
```

```python
def nn_model(X, Y, n_h, numIters = 10000, learning_rate = 1.2, print_cost = False):
    np.random.seed(3);
    n_x, n_y = layer_sizes(X, Y)

    paras = initialize_parameteres(n_x, n_h, n_y)
    print(paras["W1"].shape)
    
    for i in range (0, numIters):
        A2, cache = forward_propagation(X, paras)
        
        cost = compute_cost(A2, Y)
        
        grads = back_propagation(paras, cache, X, Y)
        
        paras = update_paras(paras, grads, learning_rate)
        
        if print_cost and i % 1000 == 0:
            print("cost after iter %i: %f" %(i, cost))
    
    return paras

    
```

```python
def update_paras(paras, grads, learning_rate=1.2):
    W1 = paras["W1"]
    b1 = paras["b1"]
    W2 = paras["W2"]
    b2 = paras["b2"]
    
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]
    
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    paras = {"W1":W1, 
            "b1":b1,
            "W2":W2,
            "b2":b2}
    
    return paras
    
    
```

```python
def back_propagation(paras, cache, X, Y):
    m = Y.shape[0]
    
    A1 = cache["A1"]
    A2 = cache["A2"]
    
    W2 = paras["W2"]
    
    assert(A2.shape == Y.shape)
    
    dZ2 = A2-Y
    dW2 = np.dot(dZ2, A1.T)/m
    db2 = np.sum(dZ2, axis=1, keepdims = True)/m
    
    dZ1 = np.multiply(np.dot(W2.T, dZ2), 1-np.power(A1, 2))
    dW1 = np.dot(dZ1, X.T)/m
    db1 = np.sum(dZ1, axis=1, keepdims = True)/m
    
    grads = {"dW1" : dW1,
             "db1" : db1,
            "dW2" : dW2,
            "db2" : db2}
    
    return grads
```

```python
def compute_cost(A2, Y):
    m = Y.shape[1]
    
    cost = -(np.dot(Y, np.log(A2.T) + np.dot(np.log(1-A2), 1-Y.T)))/m
    cost = np.squeeze(cost)
    
    return cost  
    
```

```python
def forward_propagation(X, paras):
    W1 = paras["W1"]
    b1 = paras["b1"]
    W2 = paras["W2"]
    b2 = paras["b2"]

    Z1 = np.dot(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    #y_predictions = (a2 > 0.5) * 1 
    
    n_x, n_y = layer_sizes(X, b2[0,:])
    n_h = W1.shape[0]
    m = X.shape[1]
    
    assert(Z1.shape == ((n_h, m)))
    assert(Z2.shape == ((n_y, m)))

    cache = {"Z1" : Z1,
             "A1" : A1,
             "Z2" : Z2,
             "A2" : A2}
    
    return A2, cache
    #return y_predictions
```

```python
def initialize_parameteres(n_x, n_h, n_y):
    np.random.seed(2)
    W1 = np.random.randn(n_h, n_x)*0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y, n_h)*0.01
    b2 = np.zeros((n_y, 1))
    
    parameters = {"W1":W1, 
                 "b1":b1,
                 "W2":W2,
                 "b2":b2}
    return parameters
```

```python
def layer_sizes(X, Y):
    n_x = X.shape[0];
    n_y = Y.shape[0]
    return n_x, n_y  
```
